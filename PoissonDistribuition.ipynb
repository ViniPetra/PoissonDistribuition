```python
import math
import matplotlib.pyplot as plt
```

<h1> Distribuição de Poisson </h1>
A distribuição de Poisson é a distribuição de probabilidade <strong>discreta</strong> que expressa a probablilidade de uma dada quantidade de eventos ocorram em um <strong>intervalo fixo</strong> de <strong>tempo</strong> ou <strong>espaço</strong> se esses eventos ocorrem com uma taxa média constante conhecida e <strong>independentemente</strong> do tempo desde o último evento


```python
# poisson's formula
def poisson(lam, size):
    e = 2.71828
    res = ((lam**size)*(1/(e**lam)))/math.factorial(size)
    return res

#Função de plot
def plotme(xm , ym, val):
    plt.plot(ym, xm, color = 'yellow')
    plt.bar(ym, xm, width = 0.8, color = ['red', 'green'])
    plt.xlabel('Ocorrências')
    plt.ylabel('Probabilidade')
    plt.title(val)
    return plt.show()
```

<h1>Exemplo 1</h1>
Geralmente, são feitas. em média, 4 perguntas em uma sala de aula por aula.
Qual a chance de serem feitas 7 peguntas?

```python
# Lam = 4
# Size = 7

poisson(4, 7)*100
```

<h1>Exemplo 2</h1>
Um gato traz para casa, em média, um rato morto por semana. Quais as probabilidades de este gato trazer 4, 5 ou 6 ratos?

```python
# Lam = 1
xm, ym = [], []
val = "Lambda = 1"
for size in range(10):
        x = poisson(1, size)
        xm.append(x)
        ym.append(size) 
plotme(xm, ym, val)
```

Agora, variando a média

```python
for lam in range(0,6):
    xm, ym = [], []
    val = 'Lamda'+ ' ' +str(lam)
    for size in range(10):
        x = poisson(lam, size)
        xm.append(x)
        ym.append(size) 
    plotme(xm, ym, val)
```
